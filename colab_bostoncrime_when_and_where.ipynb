```python
import urllib, requests
from collections import defaultdict

bpd_logs = "https://data.boston.gov/api/3/action/datastore_search_sql"

sql = '''
SELECT "DISTRICT", "OCCURRED_ON_DATE", "SHOOTING"
FROM "b973d8cb-eeb2-4e7e-99da-c92938efc9c0"
WHERE CAST("SHOOTING" AS INT) = 1
  AND "DISTRICT" IS NOT NULL
  AND "OCCURRED_ON_DATE" IS NOT NULL
'''

url = bpd_logs + "?sql=" + urllib.parse.quote(sql)
rows = requests.get(url).json()
records = rows['result']['records']
print(f"Fetched {len(records)} shooting records")

district_totals = defaultdict(int)
for r in records:
    district_totals[r['DISTRICT']] += 1

district_by_month = defaultdict(lambda: defaultdict(int))
for r in records:
    district = r['DISTRICT']
    month = r['OCCURRED_ON_DATE'][:7]
    district_by_month[district][month] += 1

all_months = sorted(set(
    r['OCCURRED_ON_DATE'][:7] for r in records
))

print("\nconst districtTotals =", dict(district_totals), ";")
print("\nconst allMonths =", all_months, ";")
print("\nconst districtByMonth = {")
for district, months in district_by_month.items():
    counts = [months.get(m, 0) for m in all_months]
    print(f'  "{district}": {counts},')
print("};")
```
